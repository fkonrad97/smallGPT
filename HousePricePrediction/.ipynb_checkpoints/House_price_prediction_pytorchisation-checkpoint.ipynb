```python
import torch
import pandas as pd
import numpy as np
import torch.distributions.exponential as dist
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("BostonHousing.csv")
data
```

```python
# correlation plotting: https://stackoverflow.com/questions/29432629/plot-correlation-matrix-using-pandas

data.corr().style.background_gradient(cmap='coolwarm').format(precision=2)
```

```python
data.describe()
```

```python
data.drop(columns=['chas', 'rad', 'zn'], inplace=True)
```

```python
# Removing outliers which has z-score higher than 3
''' A z-score measures exactly how many standard deviations above or below the mean a data point is. '''
# https://stackoverflow.com/questions/23199796/detect-and-exclude-outliers-in-a-pandas-dataframe?answertab=modifieddesc#tab-top

data = data[np.abs(stats.zscore(data)) < 3]
data
```

```python
# Checking for NaN values in the DataFrame

data.isna().sum()
```

```python
data = data.dropna()

data.isna().sum()
```
