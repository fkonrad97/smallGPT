```python
%pip install graphviz
from graphviz import Digraph

def trace(root):
  # builds a set of all nodes and edges in a graph
  nodes, edges = set(), set()
  def build(v):
    if v not in nodes:
      nodes.add(v)
      for child in v._prev:
        edges.add((child, v))
        build(child)
  build(root)
  return nodes, edges

def draw_dot(root):
  dot = Digraph(format='svg', graph_attr={'rankdir': 'LR'}) # LR = left to right
  
  nodes, edges = trace(root)
  for n in nodes:
    uid = str(id(n))
    # for any value in the graph, create a rectangular ('record') node for it
    dot.node(name = uid, label = "{ %s | data %.4f | grad %.4f }" % (n.label, n.data, n.grad), shape='record')
    if n._op:
      # if this value is a result of some operation, create an op node for it
      dot.node(name = uid + n._op, label = n._op)
      # and connect this node to it
      dot.edge(uid + n._op, uid)

  for n1, n2 in edges:
    # connect n1 to the op node of n2
    dot.edge(str(id(n1)), str(id(n2)) + n2._op)

  return dot
```

```python
class Value:

    # Children is to build a graph about previous calculations. Parameter given as a tuple
    def __init__(self, data, _children=(), _op='', label=''):
        self.data = data
        self.grad = 0.0 # Maintain the value of the derivative of L represented by this object Value
        self._backward = lambda: None # Default: this function does not do anything (For example for leaves nodes)
        self._prev = set(_children)
        self._op = _op
        self.label = label

    # Python uses repr funtion to print out value in a specialized way
    def __repr__(self):
        return f"Value(data={self.data}, label={self.label})"

    # For python to be able to use + for Value objects
    # # Same as to use: a.__add__(b)
    def __add__(self, other):
        out=Value(self.data + other.data, (self, other), '+')
        
        def _backward(): # Since it is addition, then it will just distribute the grad
            print('__add__: ', self.label, other.label) 
            self.grad+=1.0 * out.grad # Grads should be aggregated if it is the multivariate case. So for example it has multiple parent nodes
            other.grad+=1.0 * out.grad

        # Sets out's _backward function to the above implemented
        out._backward = _backward
        return out

    # For python to be able to use * for Value objects
    def __mul__(self, other):
        out = Value(self.data * other.data, (self, other), '*')

        def _backward(): # Since it is multiplication, then it will null out the either one of them respected
            print('__mul__: ', self.label, other.label)
            self.grad+=other.data * out.grad
            other.grad+=self.data * out.grad

        out._backward = _backward
        return out

    def __sub__(self, other): # self - other
        return Value(self.data - other.data, (self, other), '-')

    def __truediv__(self, other): # self / other
        return Value(self.data / other.data, (self, other), '/')

    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1) / (math.exp(2*x) + 1)
        out = Value(t, (self, ), 'tanh')

        def _backward():
            print('__tanh__: ', self.label, out.label)
            self.grad+=(1 - t**2) * out.grad

        out._backward = _backward
        return out

    def backward(self):
        # Topological sorting implementation:
        topo = []
        visited = set()

        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v._prev:
                    build_topo(child)
                topo.append(v)
        build_topo(self)

        self.grad=1.0
        for node in reversed(topo):
            node._backward()

    def getPrev(self):
        if self._prev != {}:
            for i in self._prev:
                print(i.label, i._op, i._prev)
                i.getPrev()
```

```python
# Let's add h to that variable which you would like to see what is the derivative to L. dL/d?
# As above: { [f(x + h) - f(x)] / h }

# Let's see the derivative of L with respect to d
# L = d * f
# dL / dd = ?
# Proof: [f(x + h) - f(x)] / h =
# [((d + h) * f) - (d * f)] / h =
# [(df + hf) - df] / h =
# hf / h = f

def mockGrad():
    h = Value(0.0001, label='h')
    
    a = Value(2.0, label='a')
    b = Value(-3.0, label='b')
    c = Value(10.0, label='c')
    e = a * b; e.label = 'e'
    d = e + c; d.label = 'd'
    f = Value(-2.0, label = 'f')
    L = f * d; L.label = 'L' 
    L1 = L

    a = Value(2.0, label='a')
    b = Value(-3.0, label='b')
    c = Value(10.0, label='c')
    e = a * b + h; e.label = 'e'
    d = e + c; d.label = 'd'
    f = Value(-2.0, label = 'f') 
    L = f * d; L.label = 'L'
    L2 = L # L + h -> will result around 1, since the derivative of L with respect to L is 1

    return ((L2-L1) / h)

mockGrad()
```

```python
# Backpropagation:
# dd / dc ?
# d = c + e
# [(c + h + e) - f(c + e)] / h = h / h = 1.0
# By symmetry to dd / de will be also 1.0
# So what is the solution? -> CHAIN RULE
# dz/dx = dz/dy * dy/dx, if z depends on y and y depends on x
# Intuition: George F. Simmons: "If a car travels twice as fast as a bicycle and the bicycle is four times as fast as a walking man, then the car travels 2 * 4 = 8 times fast as the man."
# So in this case to get to know dL / dc, we should use the chain rule by: [ dL/dc = dL/dd * dd/dc ]

# dL/dd = -2.0 / dd/dc = 1.0
c.grad = -2.0 * 1.0

# dd/de = 1.0
e.grad = -2.0 * 1.0

# dL/db = dL/dd * dd/de * de/db -> de/db = [((b + h) * a) - (b * a)] / h = (ba + ha - ba) / h = a
b.grad = -2.0 * 1.0 * 2.0

# Based on the same symmetric feature of the product here, dL/da = -3.0
a.grad = -2.0 * 1.0 * -3.0
```

```python
# Let's see a small example of using small grad change in a,b,c and f leaves nodes and let's see how L will change:
gradRate = 0.01

a = Value(2.0, label='a')
b = Value(-3.0, label='b')
c = Value(10.0, label='c')
f = Value(-2.0, label = 'f')

# Using the previously calculated grads:
a.grad = -2.0 * 1.0 * -3.0
b.grad = -2.0 * 1.0 * 2.0
c.grad = -2.0 * 1.0
f.grad = -2.0

a.data += gradRate * a.grad
b.data += gradRate * b.grad
c.data += gradRate * c.grad
f.data += gradRate * f.grad

e = a * b; e.label='e'; e.grad=-2.0
d = c + e; d.label='d'; d.grad=-2.0
L = f * d; L.label='L'; L.grad=1.0

draw_dot(L)
```

```python
# Let's create and model a neural network:

# Inputs:
x1 = Value(2.0, label='x1')
x2 = Value(0.0, label='x2')

# weights w1, w2:
w1 = Value(-3.0, label='w1')
w2 = Value(1.0, label='w2')

# Bias of the neurons:
b = Value(6.7, label='b')

# x1*w1 + x2*w2 + b
x1w1 = x1 * w1; x1w1.label='x1 * w1'
x2w2 = x2 * w2; x2w2.label='x2 * w2'
x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label='x1w1 + x2+w2'
n = x1w1x2w2 + b; n.label='n'

# Use activation function:
o = n.tanh(); o.label='output'

draw_dot(o)
```

```python
# Backpropagation part manually:
# do/do = 1
o.grad=1.0

# do/dn => o=tanh(n)
# do/dn = 1 - tanh(n)**2
# do/dn = 1 - o**2
# 1 - o.data**2 = 0.5
n.grad=1 - o.data**2

# the '+' node will only just distribute the gradient backwards and dont manipulate it, so:
x1w1x2w2.grad=0.5
b.grad=0.5

# Again a plus, so just distribute again
x1w1.grad=0.5
x2w2.grad=0.5

# Backpropagating back to the input layer, there is a multiply sign which will make disappear the respected variable, so:
x1.grad=w1.data * x1w1.grad
w1.grad=x1.data * x1w1.grad

x2.grad=w2.data * x2w2.grad
w2.grad=x2.data * x2w2.grad

draw_dot(o)
```

```python
# Let's use _backward function:

# Inputs:
x1 = Value(2.0, label='x1')
x2 = Value(0.0, label='x2')

# weights w1, w2:
w1 = Value(-3.0, label='w1')
w2 = Value(1.0, label='w2')

# Bias of the neurons:
b = Value(6.7, label='b')

# x1*w1 + x2*w2 + b
x1w1 = x1 * w1; x1w1.label='x1 * w1'
x2w2 = x2 * w2; x2w2.label='x2 * w2'
x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label='x1w1 + x2+w2'
n = x1w1x2w2 + b; n.label='n'

# Use activation function:
o = n.tanh(); o.label='output'
```

```python
# When calling the n.tahn() function, it sets the '_backward' function to 'out's, which is 'o' in this case. When 'o' calls '_backward', 
# it will run: 'self.grad=(1 - t**2) * out.grad', where self.grad has been set to 'n' and out.grad has been set to 'o's grad 
o.grad = 1.0 # Because default setting is grad=0.0
o._backward()
```

```python
# When calling the 'b + x1w1x2w2 = n', it sets the '_backward' function to 'out's, which is 'n' in this case. When 'n' calls '_backward', 
# it will run: 'self.grad=1.0 * out.grad; other.grad=1.0 * out.grad;', where self.grad has been set to 'b' and out.grad has been set to 'n's grad
# and 'other' has been set to 'x1w1x2w2'. (Or reverse)
n._backward()
```

```python
# Do the backpropagation without calling _backward always manually:

def backprop(output):
    if output._prev != {}:
        output._backward()
        for i in output._prev:
            backprop(i)

backprop(o)
```

```python
# Or using topological sort instead of using recursion: https://www.interviewcake.com/concept/java/topological-sort

# Inputs:
x1 = Value(2.0, label='x1')
x2 = Value(0.0, label='x2')

# weights w1, w2:
w1 = Value(-3.0, label='w1')
w2 = Value(1.0, label='w2')

# Bias of the neurons:
b = Value(6.7, label='b')

# x1*w1 + x2*w2 + b
x1w1 = x1 * w1; x1w1.label='x1 * w1'
x2w2 = x2 * w2; x2w2.label='x2 * w2'
x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label='x1w1 + x2+w2'
n = x1w1x2w2 + b; n.label='n'

# Use activation function:
o = n.tanh(); o.label='output'; o.grad=1.0

# Topological sorting implementation:
topo = []
visited = set()

def build_topo(v):
    if v not in visited:
        visited.add(v)
        for child in v._prev:
            build_topo(child)
        topo.append(v)
    
build_topo(o)
print(topo)

for node in reversed(topo):
    node._backward()
```
